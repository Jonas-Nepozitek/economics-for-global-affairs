```python
import pandas as pd
import numpy as np

# ---------- 1. Load full WIOD file ----------
df = pd.read_stata("WIOT2014_October16_ROW.dta")
df = df.drop(columns=["IndustryCode", "IndustryDescription", "Year", "RNr"])
```

```python
df = df.groupby("Country").sum().reset_index()
df2 = df['Country'].copy().to_frame()
```

```python
for count in df['Country']:
    newCol = count
    df2.loc[:,newCol] = df[[col for col in df.columns if col.startswith("v" + count)]].sum(axis=1)
df2 = df2.drop(columns=['TOT'])
df2 = df2[df2['Country'] != 'TOT']
```

```python
df3 = df2.iloc[:,~(df2.columns.isin(['USA','CHN','CAN','MEX']))]
df4 = df3['Country'].copy().to_frame()
df4['ROW'] = np.sum(df3.iloc[:,1:],axis = 1)
df5 = df.merge(df2.loc[:,['CAN','CHN','MEX','USA']],df4,on=['Country'])
df5
```

```python


df4[['CAN','CHN','MEX','USA']]=df2[['vCAN','vCHN','vMEX','vUSA']]
df5 = df4[df4['Country'].isin(['CAN','CHN','MEX','USA','ROW'])]
df5
```

```python
df6 = df4[~df4['Country'].isin(['CAN','CHN','MEX','USA'])]
df6 = df6.iloc[:,1:].sum(axis=0)
df6
```

```python
df5.loc[df5['Country']=='ROW'].iloc[:,1:]=df6.T
df5
```

```python
e = 5.0

l2014 = np.array([[0.923, 0.011,0.009,0.011,0.046],
                [0.004,0.941,0.0000001,0.001,0.055],
                [0.084,0.018,0.830,0.004,0.064],
                [0.090,0.015,0.006,0.832,0.057]])
l2000 = np.array([[0.932, 0.003,0.007,0.012,0.046],
                [0.004,0.931,0.0000001,0.001,0.064],
                [0.097,0.002,0.845,0.003,0.054],
                [0.114,0.004,0.005,0.816,0.061]])

t7 = np.full(shape=(4,6),fill_value=np.nan)

for i in range(0,4):
    arr = np.concatenate([np.arange(0, i), np.arange(i + 1, 5)])
    t7[i,0] = 1/e*np.sum(l2000[i,arr]/l2000[i,i]*(l2014[i,arr]-l2000[i,arr])/l2000[i,arr])

for i in range(0,4):
    for j in range(0,5):
            if j == i:
                 continue
            t7[i,j+1] = 1/e*l2000[i,j]/l2000[i,i]*(l2014[i,j]-l2000[i,j])/l2000[i,j]/t7[i,0]

df7 = pd.DataFrame(t7,columns=['GFT','US Cont','CHN Cont','MEX Cont','CAN Cont','ROW Cont'])
df7
```
