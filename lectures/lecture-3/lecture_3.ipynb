```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Lecture 2

```python
df = pd.read_csv('https://github.com/akhandelwal8/globaleconomics/blob/main/class/class_accounting.csv?raw=true', sep='\t')
```

```python
# cn = cap
# cgdpo = gdp
```

```python
df['k'] = df['cn'] / df['pop'] # calculate capital per person
df['y'] = df['cgdpo'] / df['pop'] # calculate output per person
df['A'] = df['y'] / (df['k'] ** 0.3) # calculate total factor productivity
```

```python
plt.scatter(np.log(df['y']), np.log(df['A']))
plt.xlabel('Log Output per Person (log y)')
plt.ylabel('Log Total Factor Productivity (log A)')
plt.title('Log Output per Person vs Log Total Factor Productivity')
plt.show()
```

# Lecture 3

Notes from the graph above
- Linear relationship between the two in log terms
- Stark correlation between the two
- Increasing TFP will lead to increased GDPPC, instead of just increasing capital
- Ignores other factors like energy, land, etc.
  - For countries like Qatar, this model is too simplistic

```python
plt.scatter(np.log(df['y']), np.log(df['A']))
plt.scatter(np.log(df['y']), np.log(df['k']))
plt.xlabel('Log Output per Person (log y)')
plt.ylabel('Log Total Factor Productivity (log A), Log Capital per Person (log k)')
plt.title('Log Output per Person vs Log Total Factor Productivity, Log Capital per Person')
plt.show()
```

- Countries are more elastic to TFP changes than capital changes
- Countries with the same capital per worker can have largely different gdppc
- Some countries use their capital more efficiently than others
- You can account for things like education when measuring the TFP, which could make the TFP scatterplot flatter

```python
df['k_mean'] = df['k'].mean()
df['A_mean'] = df['A'].mean()
```
