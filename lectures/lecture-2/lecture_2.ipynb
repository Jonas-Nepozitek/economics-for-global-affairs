```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('https://github.com/akhandelwal8/globaleconomics/blob/main/class/class_accounting.csv?raw=true', sep='\t')
```

```python
# cn = cap
# cgdpo = gdp
```

```python
df['k'] = df['cn'] / df['pop'] # calculate capital per person
df['y'] = df['cgdpo'] / df['pop'] # calculate output per person
df['A'] = df['y'] / (df['k'] ** 0.3) # calculate total factor productivity
```

```python
plt.scatter(np.log(df['y']), np.log(df['A']))
```
